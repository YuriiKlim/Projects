```python
from ultralytics import YOLO

model = YOLO('yolov8x.pt')

for class_id, class_name in model.names.items():
    print(f"ID: {class_id}, Class: {class_name}")
```

For parsing and collecting video frame by frame is used **ffmpeg**

Original - https://www.youtube.com/watch?v=MNn9qKG2UFI&t=51s

Output - https://youtu.be/AkG4u2nCtoA

```python
import os
import cv2

input_folder = 'C:/Users/klimk/Desktop/Highway'
output_folder = 'C:/Users/klimk/Desktop/Highway_detection'

if not os.path.exists(output_folder):
    os.makedirs(output_folder)

for filename in os.listdir(input_folder):
    if filename.endswith('.jpg') or filename.endswith('.png'):
        image_path = os.path.join(input_folder, filename)
        image = cv2.imread(image_path)

        results = model.predict(source=image,
                                conf = 0.25,
                                iou=0.9,
                                imgsz=1024)

        output_image = results[0].plot()
        output_image_path = os.path.join(output_folder, filename)
        cv2.imwrite(output_image_path, output_image)
```
